```python
import cv2
img = cv2.imread('/content/drive/MyDrive/Colab/졸업작품/EasyOCR/trainer/craft/data_root_dir/ch4_training_images/5350002-1999-0001-0001.jpg')
```

```python
import EasyOCR.easyocr.easyocr as ocr

# custom easyocr
langs = ['en']

reader = ocr.Reader(lang_list = langs, gpu = True,
                    finetuned_craft_path = 'EasyOCR/trainer/craft/exp/custom_data_train/CRAFT_clr_amp_1.pth',
                    model_storage_directory = 'deep-text-recognition-benchmark/saved_models/None-ResNet-None-CTC-Seed1111',
                    user_network_directory = 'EasyOCR/custom_network',
                    recog_network = 'custom'
                    )
```

```python
results = reader.readtext(img)
print(results)
```
